```python
import json
with open("./About2ConfigScraped.json") as file: 
	config_dict = json.load(file)
	with open('./blurbs.md', 'w') as outfile: 
		for groupName in config_dict: 
			outfile.write("\n\n")
			for config in config_dict[groupName]:
				print(config)
				outfile.write(f"---{config['key']}\n")
				front_matter = [ str(key) + ":" + str(value) + "\n" for key, value in zip(config,config.values())]
				outfile.writelines(front_matter)
				outfile.write(f"---\n\n\n")
```
